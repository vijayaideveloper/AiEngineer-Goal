```python
from collections import Counter

name = 'vijayakumar'
print(Counter(name))
print(Counter(name).items())
print(Counter(name).keys())
print(Counter(name).values())
```

```python
print(Counter(name).most_common(2))
```

```python
print(list(Counter(name).elements()))
```

```python
from collections import namedtuple

Point = namedtuple('Point', 'x,y')

pt = Point(1,-4)
print(pt)
print(pt.x, pt.y)
```

```python
from collections import deque

d = deque()
d.append(2)
d.append(3)
d.append(4)
d.append(5)
d.append(6)
d.append(7)
d.append(2)
d.append(3)
d.appendleft(1)
print(d)
d.pop()
d.popleft()
d.extend([4,5,6,7,8])
d.clear()
d
d.extend([4,5,6,7,8])
d.extendleft([44,33,322,4543])
print(d)
d.rotate(2)
d
```
