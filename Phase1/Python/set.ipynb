```python
odds = {1,3,5,7,9}
evens = {0,2,4,6,8}
primes = {2,3,5,7}

u = odds.union(evens)
u
```

```python
i = odds.intersection(evens)
i
print(i)
i2 = odds.intersection(primes)
print(i2)
i3 = evens.intersection(primes)
print(i3)
```

```python
setA = {1,2,3,4,5,6,7,8,9}
setB = {1,2,3,10,11,12}

diff1 = setA.difference(setB)
print(diff1)
diff2 = setB.difference(setA)
print(diff2)
```

```python
sym_diff1 = setA.symmetric_difference(setB)
print(sym_diff1)
sym_diff2 = setB.symmetric_difference(setA)
print(sym_diff2)
```

```python
setX = {1,2,3,4,5}
setY = {1,2,3}
setZ = {10, 20, 30}

print(setX.issubset(setY))
print(setY.issubset(setX))

print(setX.issuperset(setY))
print(setY.issuperset(setX))

print(setY.isdisjoint(setZ))
```
