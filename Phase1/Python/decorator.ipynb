```python
def decorator(func):
    def wrapper():
        print("Before Function")
        func()
        print("After Function")
    return wrapper()

def say_hi():
    print('hello')

say = decorator(say_hi)
say
```

```python
def decorator(func):
    def wrapper():
        print("Before Function")
        func()
        print("After Function")
    return wrapper()

@decorator
def say_hi():
    print('hello')

say_hi
```
